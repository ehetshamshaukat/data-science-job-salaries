```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
dataset=pd.read_csv("/kaggle/input/data-science-job-salaries/ds_salaries.csv")
dataset
```

```python
dataset.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
dataset.sample(10)
```

```python
dataset.describe()
```

```python
different_residence=dataset["employee_residence"].unique()
different_residence
num_of_different_residence=dataset["employee_residence"].nunique()
num_of_different_residence
```

```python
job_titles=dataset["job_title"].unique()
num_of_job_titles=dataset["job_title"].nunique()
num_of_job_titles
```

```python
salary_currency=dataset["salary_currency"].unique()
salary_currency
num_of_different_salary_curreny=dataset["salary_currency"].nunique()
num_of_different_salary_curreny
```

```python
experience_level=dataset["experience_level"].unique()
experience_level
```

```python
company_size=dataset["company_size"].unique()
company_size
```

```python
dataset.groupby(["work_year","experience_level","salary_currency"]).agg(minimum=("salary_in_usd","min"),maximim=("salary_in_usd","max"),average=("salary_in_usd","mean")).reset_index()
```

# By experience

```python
most_experience=dataset[dataset["work_year"]==2022]
most_experience
```

```python
most_experience=dataset[dataset["work_year"]==2020]
most_experience
```

# Employment type

```python
employment_type=dataset["employment_type"].unique()
```

```python
salary_by_employement_type=dataset.groupby("employment_type").agg(minimum=("salary_in_usd","min"),maximum_salary=("salary_in_usd","max"),average=("salary_in_usd","mean")).reset_index()
salary_by_employement_type
```

# Job title with minimum, maximum and average salary in USD

```python
dataset.groupby("job_title").agg(minimum_salary_in_usd=("salary_in_usd","min"),maximum_salary_in_usd=("salary_in_usd","max"),average_salary_in_usd=("salary_in_usd","mean")).reset_index()
```

# Remote Ratio

# Company size

```python
large_size_company=dataset.loc[dataset["company_size"]=="L"]
large_size_company
```

```python
medium_size_company=dataset.loc[dataset["company_size"]=="M"]
medium_size_company
```

```python
small_size_company=dataset.loc[dataset["company_size"]=="S"]
small_size_company
```

```python
small_size_company.groupby(["job_title","employee_residence","company_size","salary_currency"]).agg(minmum=("salary_in_usd","min"),maximum=("salary_in_usd","max")).reset_index()
```

# Company location

```python
company_location=dataset.groupby(["company_location","company_size"]).agg(num_of_company=("company_size","count")).reset_index()
company_location
```

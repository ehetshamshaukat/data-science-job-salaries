# Reuseable Code

```python
def create_pie(name,title,text):
    labels=name.index
    values=name.values
    trace0=go.Pie(values=values,labels=labels,textposition="inside",textinfo="label+percent+value",hole=0.4)
    data=[trace0]
    layout=dict(title=title,title_x=0.5,annotations=[dict(text=text,x=0.5,y=0.5,showarrow=False,font_size=14)])
    fig=dict(layout=layout,data=data)
    offline.iplot(fig)
```

```python
def create_sunburst(df,path:list,values,title):
    fig=px.sunburst(df,path=path,values=values)
    fig.update_layout(title=title,title_x=0.5)
    fig.show()
```

```python
def create_others(df,col,n,col_name,other_word):
    new_df=pd.DataFrame()
    need=df[col].value_counts(sort=True).index[:n]
    new_df[col_name]=np.where(df[col].isin(need),df[col],other_word)
    valcount=new_df[col_name].value_counts(sort=True)
    return valcount
```
